```python
tables = {x.split(os.sep)[-1].split('.')[0][3:]: get_table(x) for x in glob.glob(os.path.join(PATH, '*.txt'))}
```

```python
def get_table(path):
    colums = ['Name', 'Gender', 'Count']
    table = pd.read_csv(path, names = colums)
    return table
```

```python
def make_table(names, years, step = 1):
    table_data = {'Year' : [x for x in range(years[0], years[1] + 1, step)]}
    for name in names:
        table_data[name] = [name_sum(x, name) for x in range(years[0], years[1] + 1, step)]
    return pd.DataFrame(table_data)
```

```python
def name_sum(year, name):
    table = tables[str(year)]
    #return table
    return table.query('Name == "{}"'.format(name)).groupby('Name').sum()['Count'].values[0]
```

```python
make_table(['Ruth', 'Robert'], [1900, 2000]).plot(x = 'Year')
```

```python
def find_names(string, year):
    names = []
    for name in tables[str(year)]['Name'].values:
        if name[:len(string)] == string:
            names.append(name)
    return names
```

```python
def make_full_table(years):
    new_tables = []
    for year in range(years[0], years[1] + 1):
        new_table = tables[str(year)].apply(name_to_len, axis=1).groupby('Name').sum().reset_index()
        new_tables.append(new_table)
    return pd.concat(new_tables)
```

```python
def name_to_len(row):
    row.Name = len(row.Name)
    return row
```
